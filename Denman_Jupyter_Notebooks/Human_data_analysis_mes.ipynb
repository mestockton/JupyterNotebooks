```python
#pass in the rtData, trialData, leftKeyData, rightKeyData for a session and get the reaction times for incorrect trials
def getIncorrectRt(rtData, trialData, leftKeyData, rightKeyData):
    arr = np.where(rtData == -1)
    incorrect_trials = trialData[arr]

    allKeyData = np.concatenate((rightKeyData, leftKeyData))
    allKeyData = allKeyData.tolist()
    allKeyData.sort()

    keyPressIDX = []
    keyPressIncorrect = []
    for value in incorrect_trials:
        idx = (np.abs(allKeyData-value)).argmin()
        keyPressIDX = keyPressIDX + [idx]

    for value in keyPressIDX:
        keyPressIncorrect = keyPressIncorrect + [allKeyData[value]]

    incorrect_RTs = keyPressIncorrect-incorrect_trials
    return incorrect_RTs
```

```python
#makes a pie cart showing the % of correct and incorrect trials. Input should be rtData
def percentCorrectPie(data):
    correct = np.size(np.where(data !=- 1))
    incorrect = np.size(np.where(data == -1))
    total = np.size(data)
    sizes = [correct, incorrect]
    labels = 'Correct' , 'Incorrect'
    plt.pie(sizes, labels=labels, autopct='%1.1f%%')
    plt.title('percent correct for single trial')
    plt.show()

#pass in rtData for each session
def allSessionCorrectPie(a, **kwargs):
    b = kwargs.get('b', None)
    c = kwargs.get('c', None)
    d = kwargs.get('d', None)
    e = kwargs.get('e', None)
    correct = np.size(np.where(a !=- 1))
    incorrect = np.size(np.where(a == -1))
    for value in kwargs.values():
        correct1 = np.size(np.where(value !=- 1))
        incorrect1 = np.size(np.where(value == -1)) 
        correct = correct + correct1
        incorrect = incorrect + incorrect1
        
    sizes = [correct, incorrect]
    labels = 'Correct' , 'Incorrect'
    plt.pie(sizes, labels=labels, autopct='%1.1f%%')
    plt.title('% correct responses for all sessions')
    plt.show()    
```

```python
#shows a histogram of the reaction times for correct trials. Input should be rtData
def rtHistogram(data):
    #this removes the -1 values from rtData making corrected_rtData the same size as rewardData
    arr = np.where(data != -1)
    correct_rtData = data[arr]
    #histogram of reaction time for response for all rewarded trials
    rtPlot = plt.hist(x=correct_rtData, bins=16)
    plt.xlabel('Reaction Time (s)')
    plt.ylabel('Number of trials')
    plt.xlim(0, 2)
    plt.ylim(0,15)
    plt.title('reaction time histogram for single session')
    plt.show()

#pass in rtData for each session 
def allSessionRTHistogram(a, *args, **kwargs):
    b = kwargs.get('b', None)
    c = kwargs.get('c', None)
    d = kwargs.get('d', None)
    e = kwargs.get('e', None)
    
    arr = np.where(a != -1)
    correct_rtData1 = a[arr]
    plt.hist(correct_rtData1, bins=16, alpha=0.5, label='session', density=True)
    for value in kwargs.values():
        arr = np.where(value != -1)
        correct_rtData2 = value[arr]
        plt.hist(correct_rtData2, bins=16, alpha=0.5, label='session', density=True)

    plt.legend(loc='upper right')
    plt.xlim(0, 2)
    #plt.ylim(0,10)
    plt.title('Reaction time histogram for all sessions')
    plt.show()
    
```

```python
#pass in rtData for each session. make a bar chart for each session correct and incorrect responses
def allSessionBar(a,**kwargs):
    b = kwargs.get('b', None)
    c = kwargs.get('c', None)
    d = kwargs.get('d', None)
    e = kwargs.get('e', None)
    correct1 = np.size(np.where(a !=- 1)) 
    incorrect1 = np.size(np.where(a == -1))
    total1 = np.size(a)
    correct1 = correct1/total1
    incorrect1 = incorrect1/total1
    labels = []
    labels = ['1']
    i=2
    correct_values = [correct1]
    incorrect_values = [incorrect1]
    
    for value in kwargs.values():
        correct2 = np.size(np.where(value !=- 1)) 
        incorrect2 = np.size(np.where(value == -1))
        total2 = np.size(value)
        correct2 = correct2/total2
        incorrect2 = incorrect2/total2
        correct_values = correct_values + [correct2]
        incorrect_values = incorrect_values + [incorrect2]
        labels = labels + [str(i)]
        i = int(i)+1


    
    x = np.arange(len(labels))  # the label locations
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, correct_values, width, label='Correct')
    rects2 = ax.bar(x + width/2, incorrect_values, width, label='Incorrect')

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_ylabel('% correct')
    ax.set_title('% correct by session')
    ax.set_xticks(x)
    ax.set_xticklabels(labels)
    ax.legend()
    plt.ylim(0,1.2)
    fig.tight_layout()

    plt.show()
```

```python
#this function takes in imageData and outputs a list of floats. the float values represent the proportion of elephant(0) to cheetah(255)
def fixImageData(data):
    ''.join(data)
    newImageData = ''.join(data).split('.tif')
    newImageData = newImageData[::2]
    newImageData = newImageData[0:(len(newImageData)-1)]
    y = [z[32:] for z in newImageData]
    z = [z.replace('t', '') for z in y]
    final = [float(i) for i in z]
    return final
    

```
